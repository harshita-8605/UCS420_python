***Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports,technology, food, books, etc.)***<br>
***1. Convert text to lowercase and remove punctuation.***<br>
***2. Tokenize the text into words and sentences.***<br>
***3. Remove stopwords (using NLTK's stopwords list).***<br>
***4. Display word frequency distribution (excluding stopwords).***<br>

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter

text = """Technology is constantly evolving, transforming the way we live and work.
From smartphones to artificial intelligence, every innovation shapes our future.
I find it fascinating how technology bridges distances, connects people, and automates tasks.
It improves healthcare, education, and even entertainment.
With each passing year, new breakthroughs continue to redefine possibilities.
Technology is not just a tool—it’s a force that reshapes our world."""

text = text.lower().translate(str.maketrans('', '', string.punctuation))
words = word_tokenize(text)
sentences = sent_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered = [w for w in words if w not in stop_words]

freq = Counter(filtered)

print("Cleaned Paragraph:\n", text)
print("\nSentences:\n", sentences)
print("\nWords:\n", words)
print("\nFiltered Words:\n", filtered)
print("\nFrequency Distribution:\n")
for word, count in freq.items():
    print(f"{word}: {count}")

plt.figure(figsize=(10, 5))
plt.bar(freq.keys(), freq.values(), color='lightblue')
plt.xticks(rotation=45)
plt.title("Word Frequency")
plt.xlabel("Words")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

***Q2: Stemming and Lemmatization***<br>
***1. Take the tokenized words from Questuon 1 (after stopword removal).***<br>
***2. Apply stemming using NLTK's PorterStemmer and LancasterStemmer.***<br>
***3. Apply lemmatization using NLTK's WordNetLemmatizer.***<br>
***4. Compare and display results of both techniques.***

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer

porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

porter_stemmed = [porter.stem(w) for w in filtered]
lancaster_stemmed = [lancaster.stem(w) for w in filtered]
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]

print("Original Filtered Words:\n", filtered)
print("\nPorter Stemmer:\n", porter_stemmed)
print("\nLancaster Stemmer:\n", lancaster_stemmed)
print("\nLemmatized Words:\n", lemmatized)
```

***Q3. Regular Expressions and Text Splittng***<br>
***1. Take their original text from Question 1.***<br>
***2. Use regular expressions to:***<br>
***a. Extract all words with more than 5 letters.***<br>
***b. Extract all numbers (if any exist in their text).***<br>
***c. Extract all capitalized words.***<br>
***3. Use text spliƫng techniques to:***<br>
***a. Split the text into words containing only alphabets (removing digits and special characters).***
***b. Extract words starƟng with a vowel.***

```python
import re

original_text = """Technology is constantly evolving, transforming the way we live and work.
From smartphones to artificial intelligence, every innovation shapes our future.
I find it fascinating how technology bridges distances, connects people, and automates tasks.
It improves healthcare, education, and even entertainment.
With each passing year, new breakthroughs continue to redefine possibilities.
Technology is not just a tool—it’s a force that reshapes our world."""

words_more_than_5 = re.findall(r'\b\w{6,}\b', original_text)
numbers = re.findall(r'\d+', original_text)
capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', original_text)

alpha_words = re.findall(r'\b[a-zA-Z]+\b', original_text)
vowel_starting_words = re.findall(r'\b[aeiouAEIOU][a-zA-Z]*\b', original_text)

print("Words with more than 5 letters:\n", words_more_than_5)
print("\nNumbers in the text:\n", numbers)
print("\nCapitalized Words:\n", capitalized_words)
print("\nAlphabetic Words Only:\n", alpha_words)
print("\nWords Starting with a Vowel:\n", vowel_starting_words)
```

***Q4. Custom Tokenization & Regex-based Text Cleaning***<br>
***1. Take original text from Question 1.***<br>
***2. Write a custom tokenization function that:***<br>
***a. Removes punctuaƟon and special symbols, but keeps contractions (e.g., "isn't" should not be split into "is" and "n't").***<br>
***b. Handles hyphenated words as a single token (e.g., "state-of-the-art" remains a single token).***<br>
***c. Tokenizes numbers separately but keeps decimal numbers intact (e.g., "3.14" should remain as is).***<br>

***3. Use Regex Substitutions (re.sub) to:***<br>
***a. Replace email addresses with '<EMAIL>' placeholder.***<br>
***b. Replace URLs with '<URL>' placeholder.***<br>
***c. Replace phone numbers (formats: 123-456-7890 or +91 9876543210) with '<PHONE>' placeholder.***

```python
import re

text_q1 = """Technology is constantly evolving, transforming the way we live and work.
From smartphones to artificial intelligence, every innovation shapes our future.
I find it fascinating how technology bridges distances, connects people, and automates tasks.
It improves healthcare, education, and even entertainment.
With each passing year, new breakthroughs continue to redefine possibilities.
Technology is not just a tool—it’s a force that reshapes our world.
Contact me at john.doe@example.com or visit https://techworld.com.
You can also call +91 9876543210 or 123-456-7890 for more info."""

def custom_tokenizer(text):
    text = re.sub(r"[^a-zA-Z0-9\.\'\-\s]", "", text)
    tokens = re.findall(r"\b(?:\d+\.\d+|\w+(?:-\w+)*|'[a-z]+|\d+)\b", text)
    return tokens

tokens = custom_tokenizer(text_q1)

text_cleaned = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', '<EMAIL>', text_q1)
text_cleaned = re.sub(r'https?://\S+|www\.\S+', '<URL>', text_cleaned)
text_cleaned = re.sub(r'(\+91\s?\d{10}|\d{3}[-\s]\d{3}[-\s]\d{4})', '<PHONE>', text_cleaned)

print("Custom Tokens:\n", tokens)
print("\nCleaned Text:\n", text_cleaned)
```
