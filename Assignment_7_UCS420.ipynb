- Markdown explanaƟons for each step.
- Screenshots of the plots and outputs.<br>
Part I: Randomized Sales Data Generation (NumPy)
1. Initialize Random Seed
2. Generate Sales Data
- Create a NumPy array of random sales data for 12 months (shape = (12, 4)),
where each column represents a product category:
- Electronics, Clothing, Home & Kitchen, Sports.
- Sales values should be between 1000 and 5000 units per month.
- The range should be randomized using the seed.
3. Convert to DataFrame
- Convert the cleaned NumPy array into a Pandas DataFrame with:<br>
i. Columns: Electronics, Clothing, Home & Kitchen, Sports<br>
ii. Index: Months from Jan to Dec<br><br>
Part II: Data Manipulation and Analysis (Pandas)
1. Display the first 5 rows and summary statistics of the DataFrame.
2. Calculate the total sales per category and total sales per month.
3. Calculate the average sales growth between consecuƟve months for each category.
4. Add the following columns to the DataFrame:
- Total Sales: Sum of all categories per month.
- Growth Rate: Percentage change in Total Sales from the previous month.
5. Apply a randomized discount:
- If your roll number is even, apply a 10% discount to the Electronics category.

- If odd, apply a 15% discount to the Clothing category.
<br>

Part III: Visualizations (Matplotlib and Seaborn)
1. Plot monthly sales trends for each category using line plots.
2. Create the following plots:
- Box plots to show the sales distribution for each category.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Part I: Randomized Sales Data Generation (NumPy)
np.random.seed(123)

sales_data = np.random.randint(1000, 5000, size=(12, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, columns=categories, index=months)

# Part II: Data Manipulation and Analysis (Pandas)
df.head()
df.describe()

df.loc['Total'] = df.sum()
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100

roll_number = 123
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9
else:
    df['Clothing'] *= 0.85

df

# Part III: Visualizations (Matplotlib and Seaborn)
plt.figure(figsize=(10, 6))
for category in categories:
    sns.lineplot(x=df.index, y=df[category], label=category)
plt.title('Monthly Sales Trends per Category')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution per Category')
plt.xlabel('Category')
plt.ylabel('Sales Units')
plt.grid(True)
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])<br>
i. Find element-wise absolute value.<br>
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row.<br>
iii. Mean, Median and Standard DeviaƟon of flattened array, of each column, and
each row

```python
import numpy as np

array = np.array([[1, -2, 3],[-4, 5, -6]])

# i. Element-wise absolute value
absolute_array = np.abs(array)

# ii. Percentiles
flattened_array = array.flatten()
percentiles_flattened = np.percentile(flattened_array, [25, 50, 75])

percentiles_column = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)

# iii. Mean, Median, Std Dev
mean_flattened = np.mean(flattened_array)
median_flattened = np.median(flattened_array)
std_flattened = np.std(flattened_array)

mean_column = np.mean(array, axis=0)
median_column = np.median(array, axis=0)
std_column = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

absolute_array, percentiles_flattened, percentiles_column, percentiles_row, \
mean_flattened, median_flattened, std_flattened, \
mean_column, median_column, std_column, \
mean_row, median_row, std_row
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).<br>
i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
round_vals = np.round(a)

floor_vals, ceil_vals, trunc_vals, round_vals
```
