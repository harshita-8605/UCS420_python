 1. Create a List L that is defined as= [10, 20, 30, 40, 50, 60, 70, 80]. <br>
i. WAP to add 200 and 300 to L. <br>
ii. WAP to remove 10 and 30 from L. <br>
iii. WAP to sort L in ascending order. <br>
iv. WAP to sort L in descending order. <br>

```python
L = [10, 20, 30, 40, 50, 60, 70, 80]
L.append(200)
L.append(300)
print('List after adding 200 and 300: ')
print(L)
L.remove(10)
L.remove(30)
print('\nList after removing 10 and 30: ')
print(L)
L.sort()
print('\nList in ascending order: ')
print(L)
L.sort(reverse = True)
print('\nList in descending order: ')
print(L)
```

2. Create a tuple of marks scored as scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45) and perform the following operations using tuple functions: <br>
i. Identify the highest score and its index in the tuple. <br>
ii. Find the lowest score and count how many times it appears. <br>
iii. Reverse the tuple and return it as a list. <br>
iv. Check if a specific score '76' (input by the user) is present in the tuple and print its first occurrence index, or a message saying it's not present.

```python
scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45)

highest_score = max(scores)
highest_index = scores.index(highest_score)
print("The highest score is ", highest_score, " at index ", highest_index)

lowest_score = min(scores)
lowest_count = scores.count(lowest_score)
print("The lowest score is " ,lowest_score , "and it appears ", lowest_count, "times.")

reversed_list = list(reversed(scores))
print("The reversed tuple as a list: ", reversed_list)

specific_score = 76
if specific_score in scores:
    first_occurrence_index = scores.index(specific_score)
    print("The score ", specific_score, " is present at index ", first_occurrence_index)
else:
    print("The score ", specific_score, "is not present in the tuple.")

```

WAP to create a list of 100 random numbers between 100 and 900. Count and print
the: <br>
i. All odd numbers <br>
ii. All even numbers <br>
iii. All prime numbers

```python
import random

def check_prime(x):
    if x <= 1:
        return False
    for y in range(2, int(x**0.5) + 1):
        if x % y == 0:
            return False
    return True

nums = [random.randint(100, 900) for _ in range(100)]
odds = [n for n in nums if n % 2 != 0]
evens = [n for n in nums if n % 2 == 0]
primes = [n for n in nums if check_prime(n)]

print("Random numbers: ", nums)
print("\nTotal Odd Numbers:", len(odds))
print("Odd Numbers:", odds)
print("\nTotal Even Numbers:", len(evens))
print("Even Numbers:", evens)
print("\nTotal Prime Numbers:", len(primes))
print("Prime Numbers:", primes)
```
