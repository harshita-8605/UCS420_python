***1. Create a List L that is defined as= [10, 20, 30, 40, 50, 60, 70, 80]. <br>
i. WAP to add 200 and 300 to L. <br>
ii. WAP to remove 10 and 30 from L. <br>
iii. WAP to sort L in ascending order. <br>
iv. WAP to sort L in descending order.***

```python
L = [10, 20, 30, 40, 50, 60, 70, 80]
L.append(200)
L.append(300)
print('List after adding 200 and 300: ')
print(L)
L.remove(10)
L.remove(30)
print('\nList after removing 10 and 30: ')
print(L)
L.sort()
print('\nList in ascending order: ')
print(L)
L.sort(reverse = True)
print('\nList in descending order: ')
print(L)
```

***3. WAP to create a list of 100 random numbers between 100 and 900. Count and print
the: <br>
i. All odd numbers <br>
ii. All even numbers <br>
iii. All prime numbers***

```python
import random
list = []
for i in range(100,901):
  list.append(random.randint(100,900))
print('List: ',list)

odd_list = []
even_list = []
prime_list=[]
for i in list:
  if(i%2==0):
    even_list.append(i)
  else:
    odd_list.append(i)

for n in list:
  check = 0
  for i in range(2,n//2+1):
    if(n%i==0):
      check = check+1
  if check == 0:
    prime_list.append(n)

print('\nOdd List: ',odd_list)
print('\nEven List: ',even_list)
print('\nPrime List: ',prime_list)
```

***4. Consider the following two sets, A and B, representing scores of two teams in multiple matches. A = {34, 56, 78, 90} and B = {78, 45, 90, 23}<br>
WAP to perform the following operations using set functions:<br>
i. Find the unique scores achieved by both teams (union of sets).<br>
ii. Identify the scores that are common to both teams (intersection of sets).<br>
iii. Find the scores that are exclusive to each team (symmetric difference).<br>
iv. Check if the scores of team A are a subset of team B, and if team B's scores are a superset of team A.<br>
v. Remove a specific score X (input by the user) from set A if it exists. If not, print a message saying it is not present.***

```python
A = {34, 56, 78, 90}
B = {78, 45, 90, 23}
print('Unique scores:',A.union(B))
print('Common Scores:', A.intersection(B))
print('Exclusive scores:',A.symmetric_difference(B))
print('Is A a subset of B:',A.issubset(B))
print('Is B a superset of A:',B.issuperset(A))
n = int(input('Enter the number: '))
if(n in A):
  A.remove(n)
  print('A after removal:',A)
else:
  print('Number entered is not present')
```

***5. Write a program to rename a key city to a location in the following dictionary.***

```python
cities = {
    "New York": "USA",
    "Mumbai": "India",
    "Paris": "France",
    "Berlin": "Germany"
}

print("Original dictionary:", cities)

old = "Berlin"
new = "Frankfurt"

if old in cities:
    cities[new] = cities.pop(old)
    print('Renamed key :',old ,'to ', new)
else:
   print('Key :',old ,'is not found')

print("Updated dictionary:", cities)
```
