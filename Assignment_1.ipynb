```python
#  WAP to add all numbers divisible by 7 and 9 from 1 to n and n is given by the user.
n = int(input("Enter the value of n: "))
sum = 0
for i in range(1, n + 1):
    if i % 7 == 0 and i % 9 == 0:
        sum = sum + i
print("Sum of numbers divisible by 7 and 9 from 1 to ", n ," is: ", sum)
```

```python
# WAP to add all prime numbers from 1 to n and n is given by the user.
n = int(input("Enter the value of n: "))
sum = 0
for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        sum = sum + num

# Print the result
print("Sum of prime numbers from 1 to ", n, " is: " , sum)
```

```python
# WAP using function that add all odd numbers from 1 to n, n is given by the user.
def odd_sum(n):
    sum = 0
    for i in range(1, n + 1, 2):  # Iterate through odd numbers only
        sum = sum + i
    return sum

n = int(input("Enter the value of n: "))
print("Sum of odd numbers from 1 to " , n , " is: " , odd_sum(n))
```

```python
#  WAP using function that add all prime numbers from 1 to n, n given by the user.
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def sum_prime(n):
    sum = 0
    for i in range(2, n + 1):
        if is_prime(i):
            sum = sum + i
    return sum

n = int(input("Enter the value of n: "))
print("Sum of prime numbers from 1 to " , n , " is: " , sum_prime(n))
```
