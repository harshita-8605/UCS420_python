Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
- Generate x values from -10 to 10 using np.linspace().
- Compute y values for:<br>
o y= M⋅x^2<br>
o y=M⋅sin(x)
- Plot both functions in a single figure:<br>
o Use different colors and line styles.<br>
o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 100)

y1 = M * x**2
y2 = M * np.sin(x)

plt.plot(x, y1, 'r--', label='y = M * x^2')
plt.plot(x, y2, 'b-', label='y = M * sin(x)')

plt.title("Plot of y = Mx^2 and y = Msin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Q.2 Create a dataset of five subjects and scores.
- Convert it into a Pandas DataFrame.
- Plot the scores using a Seaborn bar plot with:<br>
o Different colors for each bar.<br>
o Annotations on top of each bar.<br>
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subjects = ['Math', 'Physics', 'Chemistry', 'English', 'Computer']
scores = [88, 76, 90, 85, 95]

df = pd.DataFrame({'Subject': subjects, 'Score': scores})

plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='Subject', y='Score', data=df, palette='pastel')

for i in range(len(df)):
    barplot.text(i, df['Score'][i] + 1, str(df['Score'][i]), ha='center', va='bottom')

plt.title('Scores in Different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y')

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
- Create a 2x2 subplot layout displaying:<br>
o Line plot showing cumulative sum.<br>
o Scatter plot with random noise.
- Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(102103456)

data = np.random.randn(50)
cumsum_data = np.cumsum(data)
noise = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(cumsum_data, color='blue')
axs[0, 0].set_title('Cumulative Sum')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Value')
axs[0, 0].grid(True)

axs[0, 1].scatter(range(50), noise, color='green')
axs[0, 1].set_title('Scatter Plot of Noise')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Noise Value')
axs[0, 1].grid(True)

axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/content/drive/My Drive/Colab Notebooks/Cognitive Computing/company_sales_data.csv'
df = pd.read_csv(file_path)

# 1. Line plot for Total profit
plt.figure(figsize=(8, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='green')
plt.title('Total Profit per Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# 2. Multiline plot for all product sales
plt.figure(figsize=(10, 6))
for column in df.columns[1:-1]:
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title('Sales Data of Different Products')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all features/attributes
plt.figure(figsize=(12, 6))
df.sum().drop('month_number').plot(kind='bar', color=sns.color_palette("pastel"))
plt.title('Total Sales/Profit for All Attributes')
plt.xlabel('Attributes')
plt.ylabel('Total Value')
plt.grid(axis='y')
plt.show()
```
