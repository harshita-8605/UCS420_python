***Q.2 Questions on Basic NumPy Array:<br>
a) Reverse the NumPy array: arr = np.array([1, 2, 3, 6, 4, 5])<br>
b) Find the most frequent value and their indice(s) in the following arrays:<br>
i. x = np.array([1,2,3,4,5,1,2,1,1,1])<br>
ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])***

```python
import numpy as np
#REVERSING
arr = np.array([1, 2, 3, 6, 4, 5])
rev = np.flip(arr)
print("After reversing: ",rev)
#MOST FREQUENT VALUE:
x = np.array([1,2,3,4,5,1,2,1,1,1])
freq = np.bincount(x).argmax()
print("Most frequent Value in x: ",freq,"And its indices: ", np.where(x == freq)[0])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])
Freq = np.bincount(y).argmax()
print("Most frequent Value in y: ",Freq,"And its indices: ", np.where(y == freq)[0])
```

***Q.4 Write program to create an 1-D NumPy array named 'Your Name' with evenly spaced 25 numbers from 10 to 100 using linspace(). Print the dimensions of the array, shape, total elements, the data type of each element and total number of bytes consumed by the array. Find the transpose of this array using reshape() attribute. Can we do the same
with T attribute?***

```python
import numpy as np
Harshita = np.linspace(10,100,num=25)
print("Array: ",Harshita)
print("\nDimention: ",Harshita.ndim)
print("\nShape: ",Harshita.shape)
print("\nTotal elements: ",Harshita.size)
print("\nData type of each element: ",Harshita.dtype)
print("\nTotal number of bytes consumed by the array: ",Harshita.nbytes)
print("\nTranspose: ",Harshita.reshape(-1,1))
print("\nApplying T: ",Harshita.T)
```

***Q5. Create a 2-D Array of three rows and four columns, named ucs420_'your_name' with following values – 10, 20, 30, 40, 50, 60, 70, 80, 90, 15, 20, 35. Compute the mean, median, max, min, unique elements. Reshape the array to four rows and three columns and
name it as reshaped_ ucs420_'your_name'. Resize the array to two rows and three columns and name it as resized_ ucs420_'your_name'.***

```python
import numpy as np
ucs420_harshita = np.array([[10,20,30,40],[50,60,70,80],[90,15,20,35]])
mean = np.mean(ucs420_harshita)
median = np.median(ucs420_harshita)
max = np.max(ucs420_harshita)
min = np.min(ucs420_harshita)
unique = np.unique(ucs420_harshita)
print("Mean: ",mean)
print("Median: ",median)
print("Max: ",max)
print("Min: ",min)
print("Unique: ",unique)
```
