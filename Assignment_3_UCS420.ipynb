***Question 1: Creating dataframe***

```python
import pandas as pd

data = {
    'Refund': ['Yes','No','No','Yes','No','No','Yes','No','No','No'],
    'Marital Status': ['Single','Married','Single','Married','Divorced','Married','Divorced','Single','Married','Single'],
    'Taxable Income (in Thousands)': [125,100,70,120,95,60,220,85,75,90],
    'Cheat': ['No','No','No','No','Yes','No','No','Yes','No','Yes']
}

df = pd.DataFrame(data)
print(df)
```

***Question 2: Locating rows 0,4,7,8***

```python
locate = df.loc[[0, 4, 7, 8]]
print(locate)
```

***Question 3: Navigate the dataframe:<br>1. Select rows from index 3 to 7<br>2. Select rows from index 4 to 8, and columns 2 to 4<br> 3. Select all rows with column index 1 to 3 (include index 3)***

```python
#rows from index 3 to 7
rows3_7 = df.iloc[3:8]
print("Rows from index 3 to 7:")
print(rows3_7)

#rows from index 4 to 8, and columns 2 to 4
rows = df.iloc[4:9, 2:5]
print("\nRows from index 4 to 8, and columns 2 to 4:")
print(rows)

#All rows with column index 1 to 3 (include index 3)
Rows = df.iloc[:, 1:4]
print("\nAll rows with column index 1 to 3 (including index 3):")
print(Rows)
```

***Question 6: Create a Sample Dataset (employees.csv)***

```python
employees_data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

employees = pd.DataFrame(employees_data)
```

***Q.6a: Shape of the DataFrame***

```python
shape = employees.shape
print(shape)  # (number of rows, number of columns)
```

***Q.6b: Summary of the DataFrame***

```python
summary = employees.info()
print(summary)
```

***Q.6c: Generate Descriptive Statistics***

```python
descriptive_stats = employees.describe()
print(descriptive_stats)
```

***Q.6d: Display First 5 and Last 3 Rows***

```python
rows5 = employees.head()
rows3 = employees.tail(3)
print(rows5)
print(rows3)
```

***Q.6e: Calculate Statistics<br>i. Average Salary:***

```python
average_salary = employees['Salary'].mean()
print(average_salary)
```

***ii. Total Bonus Paid:***

```python
total_bonus = employees['Bonus'].sum()
print(total_bonus)
```

***iii. The youngest employee's age:***

```python
youngest_age = employees['Age'].min()
print(youngest_age)
```

***iv. The highest performance rating:***

```python
highest_rating = employees['Rating'].max()
print(highest_rating)
```

***f) Sort the DataFrame by the Salary Column in Descending Order***

```python
sorted_employees = employees.sort_values(by='Salary', ascending=False)
print("\nSorted Employees by Salary:")
print(sorted_employees)
```

***g) Add a New Column for Performance Rating Categories***

```python
def categorize_rating(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

employees['Performance_Category'] = employees['Rating'].apply(categorize_rating)
print("\nEmployees with Performance Categories:")
print(employees[['Name', 'Rating', 'Performance_Category']])
```

***h) Identify Missing Values in the DataFrame***

```python
missing_values = employees.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values)
```

***i) Rename the Employee_ID Column to ID***

```python
employees.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("\nDataFrame after renaming Employee_ID to ID:")
print(employees.head())
```

***j) Find All Employees Who:<br>
i. Have More Than 5 Years of Experience:***

```python
experienced_employees = employees[employees['Years_of_Experience'] > 5]
print("\nEmployees with More Than 5 Years of Experience:")
print(experienced_employees)
```

***ii. Belong to the IT Department:***

```python
it_employees = employees[employees['Department'] == 'IT']
print("\nEmployees in the IT Department:")
print(it_employees)
```

***k) Modify the Dataset by Adding a New Column, Tax***

```python
employees['Tax'] = employees['Salary'] * 0.10
print("\nDataFrame after adding Tax column:")
print(employees[['Name', 'Salary', 'Tax']])
```
